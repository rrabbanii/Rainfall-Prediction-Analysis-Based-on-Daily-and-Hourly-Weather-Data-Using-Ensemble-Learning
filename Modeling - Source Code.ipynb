Supaya dapat diolah oleh model yang ada, dataset yang memiliki tipe data kategorik harus diubah ke dalam bentuk numerik. Pada penelitian ini, proses convert dataset kategorik tersebut dilakukan dengan menggunakan *One Hot Encoding*. 

Secara umum, pada metode ini akan diciptakan kolom baru sejumlah banyaknya nilai unik pada data dengan tipe kategorik. Metode ini dapat diimplementasikan dengan baik pada dataset dengan tipe data kategorik nominal.

Adapun metode lain yang dapat digunakan untuk mengolah data kategorik adalah *Label Encoding*. Pada metode ini, nilai pada kolom akan diubah ke dalam bentuk skala sesuai dengan banyaknya nilai unik pada fitur tersebut. Metode ini cocok diterapkan pada dataset dengan tipe data kategorik ordinal.

```python
def compute_feature_importance(voting_clf, weights):
    """ Function to compute feature importance of Voting Classifier """
    
    feature_importance = dict()
    for est in voting_clf.estimators_:
        feature_importance[str(est)] = est.feature_importances_
    
    fe_scores = [0]*len(list(feature_importance.values())[0])
    for idx, imp_score in enumerate(feature_importance.values()):
        imp_score_with_weight = imp_score*weights[idx]
        fe_scores = list(np.add(fe_scores, list(imp_score_with_weight)))
    return fe_scores
```
